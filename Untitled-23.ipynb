```python
import pandas as pd
import plotly.express as px
```

```python
crime_df = pd.read_csv('crime_dataset_india.csv')
```

```python
crime_df.head()
```

```python
crime_df['Date Reported'] = pd.to_datetime(crime_df['Date Reported'], dayfirst = True)
```

```python
crime_df['Date of Occurrence'] = pd.to_datetime(crime_df['Date of Occurrence'])
```

```python
crime_df['Time of Occurrence'] = pd.to_datetime(crime_df['Time of Occurrence'], dayfirst = True)
```

```python
crime_df['Date Case Closed'] = pd.to_datetime(crime_df['Date Case Closed'], dayfirst = True)
```

```python
crime_df.head()
```

```python
crime_df.isnull().sum()
```

```python
crime_df['Weapon Used'] = crime_df['Weapon Used'].fillna('Unknown')
```

```python
crime_df['Date Case Closed'] = crime_df['Date Case Closed'].fillna('2050-01-01')
```

```python
crime_df.isnull().sum()
```

```python
crime_city = crime_df['City'].value_counts()
```

```python
crime_city = crime_city.reset_index()
crime_city
```

```python
fig = px.bar(crime_city, x='count', y = 'City', orientation = 'h'
, title = 'Crime Count by City', labels = {'City' : 'City', 'count' : 'Number of Crimes'})
fig.update_layout(yaxis = {'categoryorder' : 'total ascending'})
fig.show()
```

The bar chart illustrates the distribution of crime counts across various Indian cities. Delhi leads with the highest number of crimes, surpassing other major cities like Mumbai and Bangalore. Hyderabad and Kolkata also show significant crime numbers, followed by cities like Chennai, Pune, and Ahmedabad. The chart indicates a steep drop in crime counts as we move down the list to smaller cities such as Faridabad and Rajkot, which have relatively lower crime figures. This visualization highlights the concentration of criminal activities in larger metropolitan areas compared to smaller cities.

```python
crime_df['crime_closure_duration'] = (crime_df['Date Case Closed'] - crime_df['Date Reported']).dt.days
crime_closure = crime_df[crime_df['Case Closed'] == 'Yes']
```

```python
fig = px.box(crime_closure, x = 'Crime Domain', y='crime_closure_duration', 
title = 'Crime Closure Duration By Crime Domain', 
labels = {'Crime Domain' : 'Crime Domain', 'crime_closure_duration' : 'Crime Closure Duration'})
fig.update_layout(xaxis = {'categoryorder' : 'total descending'})
fig.show()
```

The box plot displays the distribution of case closure times across different crime domains, measured in days.

Other Crime and Violent Crime domains exhibit wide ranges in case closure times, with many cases extending up to around 700 days. The median closure time for these categories appears to be slightly over 100 days, but there is significant variability, indicating that some cases take much longer to resolve.

Fire Accidents have a more concentrated distribution, with most cases being closed within 100 to 150 days, reflecting a more consistent and quicker resolution process.

Traffic Fatalities show the shortest case closure times, typically being resolved in under 50 days, with minimal variability compared to other crime domains.

```python
fig = px.box(crime_df, x='Crime Domain' , y='Police Deployed', 
title = 'Police Deployment Distribution by Crime Domain', 
labels = {'Crime Domain' : 'Crime Domain', 'Police Deployed' : 'Police Deployed'})
fig.update_layout(xaxis = {'categoryorder' : 'total ascending'})
fig.show()
```

The box plot visualizes the distribution of police deployment across four different crime domains: Other Crime, Violent Crime, Fire Accident, and Traffic Fatality. Each box represents the interquartile range (IQR), where the middle 50% of the data lies, and the line within each box indicates the median number of police deployed for each crime domain.

Observations:

Similar Distribution: The distributions across all four crime domains are quite similar, with medians around the 10-12 police deployment mark. Range: The whiskers extend from approximately 0 to 18-20 police deployed, suggesting that while most events fall within a common range, there are instances where the number of police deployed is significantly higher or lower. Uniformity: The consistency in distribution across different crime domains indicates that similar resources are allocated regardless of the type of incident, with no domain showing a notably higher or lower typical deployment.

```python
average_time_closure_crime_domain = crime_closure.groupby('Crime Domain')['crime_closure_duration'].mean().sort_values()
```

```python
average_time_closure_crime_domain = average_time_closure_crime_domain.reset_index()
```

```python
average_time_closure_crime_domain
```

```python
fig = px.bar(average_time_closure_crime_domain, x = 'Crime Domain', y = 'crime_closure_duration', 
title = 'Average time to close cases by crime domain', 
labels = {'Crime Domain' : 'Crime Domain', 'crime_closure_duration' : 'Average closure duration'})
fig.update_layout(xaxis = {'categoryorder' : 'total descending'})
fig.show()
```

The bar chart shows the average time required to close cases across different crime domains. The "Case Closure Duration" is measured in days.

Key observations:

Traffic Fatality: This domain has the shortest average case closure time, at around 20 days. This suggests that cases involving traffic fatalities are resolved more quickly, possibly due to the straightforward nature of these incidents or the immediate availability of evidence.

Fire Accident: Cases involving fire accidents take longer to close, with an average duration of around 50 days. This could be due to the need for detailed investigations, including forensic analysis and determining the cause of the fire.

Other Crime: This category sees an average closure duration of approximately 80 days, indicating that these cases might involve more complex investigations or legal proceedings.

Violent Crime: Violent crimes take the longest to close, with an average duration exceeding 120 days. This could reflect the complexity and severity of violent crime cases, which may require extensive investigation, evidence gathering, and legal processes.

Overall, the chart suggests that the complexity and seriousness of the crime significantly impact the time required to close cases, with violent crimes taking the longest to resolve and traffic fatalities being resolved the quickest.

```python
crime_df['Hour'] = crime_df['Time of Occurrence'].dt.hour
```

```python
crime_by_hour = crime_df.groupby('Hour').size().reset_index(name = 'Count')
```

```python
fig = px.line(crime_by_hour, x='Hour', y='Count' , 
title = 'Crime Occurrence by Time of the Day',
labels = {'Hour' : 'Hour' , 'Count' : 'Number of Crimes'})
fig.show()
```

The line chart illustrates the occurrence of crimes throughout the day, with the x-axis representing the hour of the day (from 0 to 23) and the y-axis showing the number of crimes. The pattern fluctuates, indicating that crime rates vary by time of day.

Key observations:

Crime occurrences peak around the 9th and 16th hours (approximately 9 AM and 4 PM), with significant drops during early morning hours (around 6 AM) and again late at night. The chart shows a general upward trend in crime occurrence during the day, with notable spikes and dips in various periods, suggesting times when certain crimes may be more likely to happen.
