```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing = pd.read_csv('california.csv')
```

we are going to work with California housing dataset to predict the median house price using a set of Machine learning models.


```python
housing.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
dummies = pd.get_dummies(housing['ocean_proximity'])
```

```python
label_encoder = LabelEncoder()
```

```python
for col in dummies.columns:
    dummies[col] = label_encoder.fit_transform(dummies[col])
```

```python
del dummies['ISLAND']
```

```python
del housing['ocean_proximity']
```

After getting the dummies from the categorical variable 'ocean_proximity', we encoded them into numerical dummies using LabelEcoder.

```python
data = pd.concat([housing, dummies], axis = 1)
```

```python
data.head()
```

```python
data.corr()
```

Although, some of the correlations of the independent variables with the dependent variable are low, we will test them using a linear regression model.

```python
fig, axs = plt.subplots(ncols = 2, nrows = 7, figsize = (20,10))
axs = axs.flatten()
index = 0
for k,v in data.items():
    sns.boxplot(y= k, data = data, ax = axs[index])
    index+=1
fig.delaxes(axs[13])
plt.show()
```

the boxplots show that some of the variables have some outliers.

```python
fig, axs = plt.subplots(ncols = 2, nrows = 7, figsize = (20,10))
axs = axs.flatten()
index = 0
for k,v in data.items():
    sns.histplot(y= k, data = data, ax = axs[index], kde = True)
    index+=1
fig.delaxes(axs[13])
plt.show()
```

The histograms show that longitude and latitude are bimodal, and most of the variables are most likely not normally distributed, we will transform them using log(1+x) in order to improve the precision of the models.

```python
kf = KFold(n_splits = 10)
```

using KFold technique, which makes 10 splits, each time one split is considered as a test set and computes the negative mean squared error. This technique serves as a tool to avoid overfitting.

```python
scores = cross_val_score(l_model, X, y, cv=kf, scoring = 'neg_mean_squared_error')
```

```python
print("MSE: %0.2f%% (+/- %0.2f%%)" %(scores.mean(), scores.std()))
```
