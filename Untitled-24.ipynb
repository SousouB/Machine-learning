```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
region_dummies = pd.get_dummies(marketing_data['region'])
```

```python
label_encoder = LabelEncoder()
for col in region_dummies.columns:
    region_dummies[col] = label_encoder.fit_transform(region_dummies[col])
```

```python
channel_dummies = pd.get_dummies(marketing_data['channel'])
```

```python
for col in channel_dummies.columns:
    channel_dummies[col] = label_encoder.fit_transform(channel_dummies[col])
```

```python
data = pd.concat([marketing_data.minutes_watched, marketing_data.CLV, region_dummies, channel_dummies], axis = 1)
```

We got the dummies from the categorical variables 'Region' and 'Channel' then encoded them using LabelEncoder and finally, concatenated the data to prepare it for the clustering.

```python
data.head()
```

```python
columns = ['minutes_watched', 'CLV', 'region_0', 'region_1', 'region_2', 'channel_1', 'channel_2', 'channel_3','channel_4','channel_5','channel_6','channel_7','channel_8']
```

```python
data.columns = columns
```

```python
scaler = StandardScaler()
```

```python
data = scaler.fit_transform(data)
```

```python
data = pd.DataFrame(data)
```

```python
data.columns = columns
```

```python
data.describe()
```

```python
data['minutes_watched'] = data['minutes_watched'].fillna(-1.972914e-01)
```

```python
Z = linkage(data, method = 'ward')
```

```python
plt.figure(figsize = (20,10))
dendrogram(Z, truncate_mode = 'level', p =5, show_leaf_counts = False, no_labels = True)
plt.title('Dendrogram')
plt.xlabel('Observation')
plt.ylabel('Distance')
plt.show()
```

The dendrogram outlines that there will be 8 clusters in our dataset.

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 8, init = 'k-means++', random_state = 42)
```

```python
km.fit(data)
```

```python
labels = km.labels_
```

```python
marketing_data['Labels'] = labels
```

```python
centers = km.cluster_centers_
```

```python
plt.figure(figsize = (20,10))
plt.scatter(data['minutes_watched'], data['CLV'], c = labels, marker = 'o' )
plt.scatter(centers[:,0], centers[:,1], marker = '*')
plt.show()
```

```python
result = marketing_data.groupby('Labels').mean()
```

```python
index = ['Segment0', 'Segment1', 'Segment2', 'Segment3', 'Segment4', 'Segment5', 'Segment6', 'Segment7']
result.index=index
```
